### Helper functions

```python
def get_state(cities, india_map):
    state_list = []
    for city in cities:
        for key, values in india_map.items():
            if city in values:
                state_list.append(key)
    if state_list:
        return list(set(state_list))
    else:
        return list(set(cities))
```

### Describe data

```python
df = pd.read_csv('marketing_sample_for_naukri_com-jobs__20190701_20190830__30k_data.csv')
state_cities = pd.read_csv('list_of_cities_and_towns_in_india.csv')
```

### Locations

```python
# Get city mapping
state_cities = state_cities.apply(lambda x: x.astype(str).str.upper())
india_map = {}
for state in state_cities['State'].unique():
    india_map[state] = state_cities.loc[state_cities.State == state, 'Name of City'].unique().tolist()

# Convert locations to upper and remove extra characters
df['location_processed'] = df['Location'].apply(lambda x: str(x).upper().split(','))
df['location_processed'] = df['location_processed'].apply(lambda x: [elements.strip() for elements in x])

# Get state of the city
df['state'] = df['location_processed'].apply(lambda x: get_state(x, india_map))

# Expand list of cities to individual rows
cols = list(df.columns)
cols.remove('state')
df_locations = pd.DataFrame({
                              col:np.repeat(df[col].values, df['state'].str.len())
                              for col in df.columns.drop('state')}
                            ).assign(**{'state':np.concatenate(df['state'].values)})

df_locations['is_real_state'] = df_locations['state'].apply(lambda x: True if x in india_map.keys() else False)
```
