```python
for cols in num_cols:
    # Overall
    print(data[cols].describe())
    
    # Hist plot
    fig = px.histogram(data, 
                       x=cols)
    fig.update_layout(bargap=0.1,
                      title=dict(text='Distribution of {}'.format(cols),
                                 x=0.5),
                     height=400,
                     width=1000)
    fig.show()
```
